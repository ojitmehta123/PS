```python
import tkinter as tk
from tkinter import ttk
import matplotlib
matplotlib.use("TkAgg")
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk
from matplotlib.figure import Figure
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
class Gui_ntc(object):
    def __init__(self):
        #init variables
        self.root=tk.Tk() #root --> self.root
        self.root.geometry("600x350")
        self.root.title("NTCL sales")       
        self.photo = mpimg.imread("NTCL_Logo.gif")
       
        f = Figure()
        f.figimage(self.photo, resize=True)
        self.canvas = FigureCanvasTkAgg(f , self.root)
        self.toolbar=NavigationToolbar2Tk(self.canvas , self.root)
        
        self.root_bg()
        
    def root_bg(self, button1 = None):
        self.canvas.get_tk_widget().destroy()
        f = Figure()
        f.figimage(self.photo, resize=True)
        self.canvas = FigureCanvasTkAgg(f , self.root)
        self.canvas.draw()
        self.canvas.get_tk_widget().pack(expand=True)
        try :
            button1.destroy()
        except :
            pass
       
    def start_gui(self):
        self.root.mainloop()


class Page(Gui_ntc):
    def __init__(self , *args):
        Gui_ntc.__init__(self)
        self.raw_frame = tk.Frame(self.root)
        self.menu = tk.Menu(self.root) # main menu     
        self.mill_menu = tk.Menu(self.menu , tearoff=0)
        self.data_menu = tk.Menu(self.menu , tearoff=0)
        self.int_var = tk.StringVar() # for mill menu variable
        self.root.config(menu=self.menu)
        
        self.plots_mill=plots_mill.copy()
        
        self.button1 = tk.Button(self.root, text="Home",
                    command=lambda: self.root_bg(self.button1))
#         self.button1.pack(side=tk.BOTTOM)

        self.init_mill_menu()
        self.init_data_menu()        
        

    def  init_mill_menu(self):
        """Set up the Mill Menu"""
        self.menu.add_cascade(label="Mill", menu=self.mill_menu)
        for i in range(13):
            self.mill_menu.add_radiobutton(label=list(self.plots_mill.keys())[i] , var = self.int_var , command = lambda : self.mill_cb())
        
    def mill_cb(self):
        for f in (self.button1 , self.toolbar , self.canvas.get_tk_widget() , self.raw_frame):
            try:
                f.destroy()
            except:
                pass

        self.button1 = tk.Button(self.root, text="Home",
                    command=lambda: self.root_bg(self.button1))
        self.button1.pack(side=tk.BOTTOM)
        self.canvas=FigureCanvasTkAgg(self.plots_mill[str(self.int_var.get())] , self.root)
        self.toolbar=NavigationToolbar2Tk(self.canvas , self.root)
        self.canvas.draw()
        self.canvas.get_tk_widget().pack(side=tk.TOP , fill=tk.BOTH , expand=True)
        
    def init_data_menu(self):
        self.menu.add_cascade(label="Data", menu=self.data_menu)
        self.data_menu.add_radiobutton(label="Raw" , command=lambda : self.mill_raw())
        
    def mill_raw(self):
        for f in (self.button1 , self.toolbar , self.canvas.get_tk_widget(),self.raw_frame):
            try:
                f.destroy()
            except:
                pass
        
        x=PrettyTable()
        column_names=['Year','Rate','Bags']
        x.add_column(column_names[0],['14-15','15-16','16-17','17-18'])
        x.add_column(column_names[1],rate_raw[str(self.int_var.get())])
        x.add_column(column_names[2],bags_raw[str(self.int_var.get())])
        
        self.raw_frame = tk.Frame(self.root)
        self.raw_frame.pack(expand=True)
        tk.Label(self.raw_frame , text=str(self.int_var.get()) , font=("Roman",24,'bold')).pack(side=tk.TOP)
        msg= tk.Message(master=self.raw_frame , text=str(x))
        msg.config(font=('roman', 24 ))
        msg.pack()
        self.raw_frame.tkraise()
    

        
```

```python
class root_class(tk.Tk):
    def __init__(self, *args):
        tk.Tk.__init__(self)
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        container.grid_columnconfigure(0, weight=1)
        container.grid_rowconfigure(0, weight=1)
        self.plots_mill = plots_mill.copy()

        self.frames = {}

        for F in (StartPage,):  # , GraphPage , DataPage):
            frame = F(container, self)
            self.frames[F] = frame
            frame.grid(row=0, column=0, sticky="nsew")

        self.show_frame(StartPage)
        self.frames[GraphPage] = GraphPage(container, self)
        GraphPage.grid(self,row=0, column=0, sticky="nsew")

    def show_frame(self, i):
        f = self.frames[i]
        f.tkraise()
        

class StartPage(tk.Frame):
    def __init__(self, r_frame, controller):
        tk.Frame.__init__(self, r_frame)
        self.photo = mpimg.imread("NTCL_Logo.gif")
        self.show_img()
        self.menu = tk.Menu(self)  # main menu
        self.mill_menu = tk.Menu(self.menu, tearoff=0)
        self.int_var = tk.StringVar()  # for mill menu variable
        controller.config(menu=self.menu)
        self.plots_mill = plots_mill.copy()
        self.init_mill_menu()

    def show_img(self):
        f = Figure()
        f.figimage(self.photo, resize=True)
        canvas = FigureCanvasTkAgg(f, self)
        canvas.draw()
        canvas.get_tk_widget().pack(expand=True)

    def init_mill_menu(self):
        """Set up the Mill Menu"""
        self.menu.add_cascade(label="mill", menu=self.mill_menu)
        for i in range(13):
            self.mill_menu.add_radiobutton(label=list(self.plots_mill.keys())[i],
                                           var=self.int_var,
                                           command=lambda:
                                           controller.frames[GraphPage].draw_graph(self.plots_mill[str(self.int_var.get())]))


class GraphPage(tk.Frame):
    def __init__(self, r_frame, controller):
        tk.Frame.__init__(self)
        button1 = tk.Button(self, text="Home",
                            command=lambda: controller.show_frame(StartPage))
        
        button1.pack()
#         self.toolbar.destroy()
#         self.canvas.get_tk_widget().destroy()
        def draw_graph(self , plots_f):
            self.canvas = FigureCanvasTkAgg(plots_f, self)
            self.canvas.draw()
            self.canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=True)

            self.toolbar = NavigationToolbar2Tk(self.canvas, self.root)
            self.toolbar.update()
            self.canvas.get_tk_widget().pack(side=tk.TOP, fill=tk.BOTH, expand=True)
            self.tkraise()

class DataPage(tk.Frame):
    pass


app = root_class(plots_mill)
app.mainloop()
```
