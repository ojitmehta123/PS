```python
df_sales = pd.read_excel(io="Sales Data.xls",header=1,index_col='Mill Name')
df_production = pd.read_excel(io="Production Data.xls",header=1,index_col='Mill Name')
```

```python
df_sales_enc=df_sales.assign(date_cat = lambda row :
                       row["Enquiry Date"].str.split(' ').str.get(0))
df_sales_enc.drop(labels=[ 'Enquiry No' , 'Order No' , 'Enquiry Date',
                      'Status','Discount Value','GST Value',
                     'Net Ex-Mill Invoice Value',
                     'Total Invoice Value' , 'Ex-Mill Invoice Value'] , axis=1)


df_sales_enc['Date']=pd.to_datetime(df_sales_enc["date_cat"])

df_sales_enc.rename(index={'A':'Burhanpur Tapti Textile Mills',
                             'B':'Aarti Cotton Textile Mills',
                             'C':'Barshi Textile Mills',
                            'D':'Finlay Mill Achalpur',
                             'E':'Finlay Mill Mumbai',
                             'F':'India United Mills No5',
                            'G':'New Bhopal Textile Mills',
                             'H':'Podar Mills',
                             'I':'Rajnagar Textile Mills',
                            'J':'Tata Mills'},inplace=True)
df_sales_enc.reset_index(inplace=True)
df_sales_enc.set_index("Date" , inplace=True)
```

```python
df_prod_enc=df_production.dropna(how='any')
df_prod_enc["Date"]=pd.to_datetime(df_prod_enc["Date"])
df_prod_enc.reset_index(inplace=True)
df_prod_enc.set_index("Date" , inplace=True)
```

```python
def set_sales_graph(mill_name , variety , df_sales_enc):
# Making sales graph from top df_sales_enc and df_prod_enc
        if mill_name is not "All":
                temp_a1 = df_sales_enc["Mill Name"]==mill_name
        else:
                temp_a1 = np.ones((len(df_sales_enc["Mill Name"])),dtype=bool)

        if variety is not "All":
                temp_a2 = df_sales_enc["Variety Name"]==variety
        else:
                temp_a2 = np.ones((len(df_sales_enc["Variety Name"])),dtype=bool)
        _df_sales_enc = df_sales_enc[np.logical_and(temp_a1,temp_a2)]
        print(type(_df_sales_enc))
        _df_sales_enc.plot(y="Rate",fontsize=20)
        _df_sales_enc.plot(y="Total Bags",fontsize=20)
```

```python
set_sales_graph("Aarti Cotton Textile Mills" , "All" , df_sales_enc)
```
