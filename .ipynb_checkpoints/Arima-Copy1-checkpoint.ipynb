```python
a=pd.date_range('2019-03-01', '2019-06-01', freq='MS')
```

```python
from pyramid.arima import auto_arima


def best(m,train,valid):
    train=train
    valid=valid
    future=a
    model = auto_arima(train, start_p=24, start_q=24,start_Q=12,start_P=12,max_P=48,max_Q=48,max_p=48, max_q=48,m=m, max_order=96,seasonal=True,d=1, D=1, trace=True,error_action='ignore',suppress_warnings=True)
    model.fit(train)

    forecast = model.predict(n_periods=len(valid)+len(a))
    forecast = pd.DataFrame(forecast,index = np.append(valid.index ,a) ,columns=['Prediction'])
    return forecast
```

```python
def mill(mil):   
    bes_m=0
    best_rms=10000
    best_for=0
    for i in range(2,25):
        train = mil[:36]
        valid = mil[36:]
        forc=best(i,train,valid)
        rms=np.sqrt(np.mean(np.power((np.array(valid)-np.array(forc['Prediction'][:len(valid)])),2)))
        if rms<best_rms:
            best_for=forc
            bes_m=i
            best_rms=rms
    print((str)(bes_m)+" "+(str)(best_rms))        
    plt.figure(figsize=(12,12))
    plt.plot(train)
    plt.plot(valid)
    plt.plot(best_for['Prediction'])
    plt.show
```

```python
mill(BT_54)
```

```python
mill(RTM_40k)
```

```python
mill(Tata_60P)
```
